```python
import numpy as np
import matplotlib.pyplot as plt

def intensity(x, a, wavelength):
    """
    Calcula a intensidade de difração de Fraunhofer para uma abertura circular.

    :param x: Variável relacionada à posição no espaço de Fourier.
    :param a: Raio da abertura circular (em metros).
    :param wavelength: Comprimento de onda da luz (em metros).
    :return: Intensidade de difração de Fraunhofer.
    """
    alpha = np.pi * a * (1/wavelength) * x
    return ( np.sin(alpha) / alpha ) ** 2

def plot_intensity(x_values, intensity_values):
    """
    Plota o gráfico da intensidade de difração.

    :param x_values: Vetor de valores da variável x.
    :param intensity_values: Vetor de valores da intensidade de difração.
    """
    plt.plot(x_values, intensity_values, label = 'Fenda Simples (a = 40\u00B5m) ')
    plt.xlabel('x (m)')
    plt.ylabel('Intensidade de Difração')
    plt.title('Intensidade de Difração de Fraunhofer')
    plt.grid(True)
    plt.legend()
    plt.show()

# Parâmetros

a = 40e-6  # Raio da abertura circular (em metros)
wavelength = 632.8e-9  # Comprimento de onda da luz (em metros)

# Vetor de valores de x
x_values = np.linspace(-0.0874, 0.09039, 1610) #ver o intervalo nos dados obtidos pelo Origin

# Calcula a intensidade de difração correspondente
intensity_values = intensity(x_values, a, wavelength)

# Plot
plot_intensity(x_values, intensity_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def intensity_single_slit(x, a, wavelength):
    """
    Calcula a intensidade de difração de Fraunhofer para uma abertura circular.

    :param x: Variável relacionada à posição no espaço de Fourier.
    :param a: Raio da abertura circular (em metros).
    :param wavelength: Comprimento de onda da luz (em metros).
    :return: Intensidade de difração de Fraunhofer.
    """
    alpha = np.pi * a * (1/wavelength) * x
    return (np.sin(alpha) / alpha) ** 2

def intensity_double_slit(x, a, wavelength, d):
    """
    Calcula a intensidade de difração de Fraunhofer para duas fendas.

    :param x: Variável relacionada à posição no espaço de Fourier.
    :param a: Raio da abertura circular (em metros).
    :param wavelength: Comprimento de onda da luz (em metros).
    :param d: Distância entre as fendas (em metros).
    :return: Intensidade de difração de Fraunhofer.
    """
    alpha = np.pi * a * (1/wavelength) * x
    gamma = np.pi * d * (1/wavelength) * x
    return (np.sin(alpha) / alpha) ** 2 * np.cos(gamma) ** 2

def plot_intensity(x_values, intensity_values_single, intensity_values_double_1, intensity_values_double_2):
    """
    Plota o gráfico da intensidade de difração.

    :param x_values: Vetor de valores da variável x.
    :param intensity_values_single: Vetor de valores da intensidade de difração para uma fenda.
    :param intensity_values_double_1: Vetor de valores da intensidade de difração para duas fendas com d = 500µm.
    :param intensity_values_double_2: Vetor de valores da intensidade de difração para duas fendas com d = 250µm.
    """
    plt.plot(x_values, intensity_values_single, label='Fenda Simples (a = 40µm)')
    plt.plot(x_values, intensity_values_double_1, label='Fenda Dupla (d = 500µm)', color='orangered')
    plt.plot(x_values, intensity_values_double_2, label='Fenda Dupla (d = 250µm)', color='moccasin' )
    plt.xlabel('x (m)')
    plt.ylabel('Intensidade de Difração')
    plt.title('Intensidade de Difração de Fraunhofer')
    plt.grid(True)
    plt.legend()
    plt.show()

# Parâmetros
a = 40e-6  # Raio da abertura circular (em metros)
wavelength = 632.8e-9  # Comprimento de onda da luz (em metros)
d_double_1 = 500e-6 # Distância entre as fendas para o primeiro caso de fenda dupla
d_double_2 = 250e-6 # Distância entre as fendas para o segundo caso de fenda dupla

# Vetor de valores de x
x_values = np.linspace(-0.0874, 0.09039, 1610)  # ou utilize o intervalo que preferir

# Calcula as intensidades de difração correspondentes
intensity_values_single = intensity_single_slit(x_values, a, wavelength)
intensity_values_double_1 = intensity_double_slit(x_values, a, wavelength, d_double_1)
intensity_values_double_2 = intensity_double_slit(x_values, a, wavelength, d_double_2)

# Plot
plot_intensity(x_values, intensity_values_single, intensity_values_double_1, intensity_values_double_2)
```
